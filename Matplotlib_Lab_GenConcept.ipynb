```python
import matplotlib.pyplot as plt 
import numpy as np
fig = plt.figure() # an empty figure with no axes
fig.suptitle('No axes in this figure') # add title name
fig, ax_lst_1 = plt.subplots(2,2) # add figure size
fig, ax_lst_2 = plt.subplots(3,2)
```

```python
x = np.linspace(0,2,100)

plt.plot(x, x, label = 'Linear')
plt.plot(x, x**2, label = 'Quardratic')
plt.plot(x, x**3, label = 'Cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Test Graph')

plt.legend()

plt.show()
```
